```python
from transformers import pipeline
#this pipelines gives every model which transformer having
#tools that help us to perform the series of steps
```

```python
classifer = pipeline ("sentiment-analysis")
#one of the model with the sentiment analysis but default model, we have not assingned any specific model
```

```python
classifer("Food is delicious!")
#checking the sentiment
```

```python
classifer("I wish food was good!")
```

```python
classifer("I wish food was good")
```

```python
classifer("I'm amazed by the speed of the processor but disappointed that it heats up quickly.")
```

```python
print(classifer(["I've been waiting for a HuggingFace course my whole life.", "I Really excited but not wanted this so much!", "Hated this"]))
```

```python
classifer( ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"])
```

```python
zero_shot = pipeline("zero-shot-classification")
#Zero-shot classification is a machine learning technique that allows a model to classify text or images into categories without being trained
#on examples from those categories
```

```python
zero_shot("This course is about generative AI and we are learning aboout Huggingface",
          candidate_labels=['education','politics','business']
          )
```

```python
zero_shot("India won 2 bronze medals in Pistol shooting",
          candidate_labels=['education','sports','business']
          )
```
